```python
for filename in glob.glob('/data/user/wsmonroe/Opthamology/neuralNetTest/gtdata/masks/*.tif'): #assuming gif
    im=Image.open(filename)
    im=im.resize((256,256))
    immax=np.max(np.max(im))
    immin=np.min(np.min(im))
    print(immax)
    print(immin)
    # Pixel range is 0...255, 256/2 = 128
    # Let numpy do the heavy lifting for converting pixels to pure black or white
    bw = np.asarray(im).copy()

    # Pixel range is 0...255, 256/2 = 128
    bw[bw < ((immax-immin)/2)+immin] = 0    # Black
    bw[bw >= ((immax-immin)/2)+immin] = 255 # White

    # Now we put it back in Pillow/PIL land
    imfile = Image.fromarray(bw)

    #im=scipy.m.misc.imresize(im,new_shape)
    #imtest = im.reshape()
    #image_list.append(im)
    print(filename)
    #print(im.shape)
    justTheName=(os.path.basename(filename))
    imfile.save(os.path.join('/data/user/wsmonroe/Opthamology/neuralNetTest/gtdata/masks/',justTheName), "TIFF")
    print(np.max(np.max(imfile)))
    print(np.min(np.min(imfile)))
```
