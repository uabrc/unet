```python
for filename in glob.glob('/data/user/wsmonroe/MetalsGroup/training_set/*_Annotation.png'): #assuming gif
    im=Image.open(filename)
    im=im.resize((512,512))
    # Pixel range is 0...255, 256/2 = 128
    # Let numpy do the heavy lifting for converting pixels to pure black or white
    bw = np.asarray(im).copy()

    # Pixel range is 0...255, 256/2 = 128
    bw[bw < 128] = 0    # Black
    bw[bw >= 128] = 255 # White

    # Now we put it back in Pillow/PIL land
    imfile = Image.fromarray(bw)

    #im=scipy.m.misc.imresize(im,new_shape)
    #imtest = im.reshape()
    #image_list.append(im)
    print(filename)
    #print(im.shape)
    justTheName=(os.path.basename(filename))
    imfile.save(os.path.join('/data/user/wsmonroe/MetalsGroup/resizedTraining/labels/masks',justTheName), "PNG")
    print(np.max(np.max(imfile)))
    print(np.min(np.min(imfile)))
```
